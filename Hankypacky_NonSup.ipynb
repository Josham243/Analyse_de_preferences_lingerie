```python
datas =pd.get_dummies(dataframe, columns=['product_category'], drop_first=True)
datas.head()
```

```python
cols = datas.columns
cols
```

```python
X_train = datas.iloc[:, 0:19].values
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scale = scaler.fit_transform(X_train)
X_scale.shape
```

```python
X_train = pd.DataFrame(X_scale, columns =[cols])
X_train.shape
```

```python
#determination du nombre de k(classes) avec la methodes du coude(Eblow en anglais)
from sklearn.cluster import KMeans

tab = []

for i in range (1,5):
    kmeans = KMeans(n_clusters = i)
    kmeans = kmeans.fit(X_scale)
    tab.append(kmeans.inertia_)
```

```python
#Visualiser Elbow
plt.plot(range(1,5), tab)
plt.title("La methode eblow")
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertie")    
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score

kmeans = KMeans(n_clusters=2)
kmeans.fit(X_scale)
label = kmeans.predict(X_scale)

print("Silhouette Score(n=2):", silhouette_score(X_scale, label) )
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans(random_state=120)
#Instancie le KElbowVisualizer avec le nombre de clusters et la métrique

visualizer = KElbowVisualizer(model, metric='silhouette', timings=False)
plt.figure(figsize=(8,5))

visualizer.fit(X_scale)
visualizer.poof()
```
